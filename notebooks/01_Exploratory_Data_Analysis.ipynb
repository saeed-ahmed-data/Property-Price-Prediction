```python
#  1: Upload dataset
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visualization preferences
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

print("ðŸ“¤ Click 'Choose Files' below to upload your train.csv file")
print("=" * 60)

uploaded = files.upload()

print("\nâœ… File uploaded successfully!")
```

```python
#  2: Load and first look at the data
df = pd.read_csv('train.csv')

print("âœ… DATASET LOADED SUCCESSFULLY!")
print("=" * 70)
print(f"\nðŸ“Š DATASET SHAPE:")
print(f"   â€¢ Total Records (Houses): {df.shape[0]:,}")
print(f"   â€¢ Total Features (Columns): {df.shape[1]}")

print(f"\nðŸ’° SALE PRICE STATISTICS:")
print(f"   â€¢ Minimum Price: ${df['SalePrice'].min():,.0f}")
print(f"   â€¢ Maximum Price: ${df['SalePrice'].max():,.0f}")
print(f"   â€¢ Average Price: ${df['SalePrice'].mean():,.0f}")
print(f"   â€¢ Median Price: ${df['SalePrice'].median():,.0f}")
print(f"   â€¢ Standard Deviation: ${df['SalePrice'].std():,.0f}")

print(f"\nðŸ“‹ FIRST 5 ROWS:")
print("=" * 70)
df.head()
```

```python
#  3: Data types and basic info
print("ðŸ“‹ DATASET INFORMATION:")
print("=" * 70)
df.info()

print("\n\nðŸ“Š SUMMARY STATISTICS:")
print("=" * 70)
df.describe()
```

```python
#  4: Check for missing values
print("ðŸ” MISSING DATA ANALYSIS:")
print("=" * 70)

# Calculate missing values
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

if len(missing) > 0:
    missing_percent = (missing / len(df) * 100).round(2)
    missing_df = pd.DataFrame({
        'Missing Count': missing,
        'Percentage': missing_percent
    })

    print(f"\nâš ï¸  Found {len(missing)} columns with missing values:\n")
    print(missing_df.head(20))

    print(f"\nðŸ“Š Total missing values: {missing.sum():,}")
else:
    print("âœ… No missing values found!")
```

```python
#  5: Visualize missing data
import matplotlib.pyplot as plt

if len(missing) > 0:
    plt.figure(figsize=(14, 8))
    missing_top = missing.head(20)

    plt.barh(range(len(missing_top)), missing_top.values, color='coral')
    plt.yticks(range(len(missing_top)), missing_top.index)
    plt.xlabel('Number of Missing Values', fontsize=12)
    plt.title('Top 20 Columns with Missing Values', fontsize=14, fontweight='bold')
    plt.gca().invert_yaxis()

    # Add value labels
    for i, v in enumerate(missing_top.values):
        plt.text(v + 10, i, str(v), va='center', fontsize=9)

    plt.tight_layout()
    plt.show()
else:
    print("No missing values to visualize!")
```

```python
#  6: Analyze and visualize price distribution
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histogram
axes[0].hist(df['SalePrice'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Sale Price ($)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].set_title('Distribution of House Prices', fontsize=14, fontweight='bold')
axes[0].axvline(df['SalePrice'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: ${df["SalePrice"].mean():,.0f}')
axes[0].axvline(df['SalePrice'].median(), color='green', linestyle='--', linewidth=2, label=f'Median: ${df["SalePrice"].median():,.0f}')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Box plot
axes[1].boxplot(df['SalePrice'], vert=True)
axes[1].set_ylabel('Sale Price ($)', fontsize=12)
axes[1].set_title('Box Plot of House Prices', fontsize=14, fontweight='bold')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Print insights
print("ðŸ’¡ PRICE DISTRIBUTION INSIGHTS:")
print("=" * 70)
print(f"â€¢ The distribution appears {'right-skewed' if df['SalePrice'].skew() > 0 else 'left-skewed'}")
print(f"â€¢ Skewness: {df['SalePrice'].skew():.2f}")
print(f"â€¢ Most houses are priced around ${df['SalePrice'].median():,.0f}")
```

```python
#  7: Identify and analyze numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_cols.remove('Id')  # Remove ID column
if 'SalePrice' in numerical_cols:
    numerical_cols.remove('SalePrice')  # We'll analyze this separately

print(f"ðŸ“Š NUMERICAL FEATURES:")
print("=" * 70)
print(f"Total numerical features: {len(numerical_cols)}")
print(f"\nList of features: {', '.join(numerical_cols[:10])}...")

# Summary statistics for key numerical features
key_features = ['LotArea', 'YearBuilt', 'GrLivArea', 'TotalBsmtSF', 'GarageArea']
print(f"\nðŸ“ˆ KEY NUMERICAL FEATURES SUMMARY:")
print("=" * 70)
df[key_features].describe()
```

```python
#  8: Calculate correlation with SalePrice
correlations = df[numerical_cols].corrwith(df['SalePrice']).sort_values(ascending=False)

print("ðŸ”— TOP 15 FEATURES CORRELATED WITH SALE PRICE:")
print("=" * 70)

top_corr = correlations.head(15)
for feature, corr in top_corr.items():
    bar = 'â–ˆ' * int(abs(corr) * 20)
    print(f"{feature:20s} {corr:6.3f} {bar}")

# Visualize correlations
plt.figure(figsize=(12, 8))
top_corr.plot(kind='barh', color='teal')
plt.xlabel('Correlation with Sale Price', fontsize=12)
plt.title('Top 15 Features Most Correlated with Sale Price', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#  9: Create correlation heatmap
# Select top correlated features for better visualization
top_features = correlations.head(10).index.tolist() + ['SalePrice']
correlation_matrix = df[top_features].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix,
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            square=True,
            linewidths=1,
            cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap: Top Features vs Sale Price', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

print("\nðŸ’¡ CORRELATION INSIGHTS:")
print("=" * 70)
print("â€¢ Red colors = Strong positive correlation")
print("â€¢ Blue colors = Strong negative correlation")
print("â€¢ White/neutral = Weak or no correlation")
```

```python
#  10: Scatter plots for top correlated features
top_4_features = correlations.head(4).index.tolist()

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for idx, feature in enumerate(top_4_features):
    axes[idx].scatter(df[feature], df['SalePrice'], alpha=0.5, color='steelblue', s=30)
    axes[idx].set_xlabel(feature, fontsize=12)
    axes[idx].set_ylabel('Sale Price ($)', fontsize=12)
    axes[idx].set_title(f'Sale Price vs {feature}', fontsize=13, fontweight='bold')
    axes[idx].grid(alpha=0.3)

    # Add trend line
    z = np.polyfit(df[feature].fillna(df[feature].median()), df['SalePrice'], 1)
    p = np.poly1d(z)
    axes[idx].plot(df[feature], p(df[feature]), "r--", linewidth=2, alpha=0.8)

plt.tight_layout()
plt.show()

print(f"\nðŸ’¡ RELATIONSHIP INSIGHTS:")
print("=" * 70)
for feature in top_4_features:
    corr = correlations[feature]
    print(f"â€¢ {feature}: {corr:.3f} correlation - {'Strong positive' if corr > 0.5 else 'Moderate positive'} relationship")
```

```python
#  11: Analyze categorical features
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f"ðŸ“‹ CATEGORICAL FEATURES:")
print("=" * 70)
print(f"Total categorical features: {len(categorical_cols)}")
print(f"\nFeatures: {', '.join(categorical_cols[:10])}...")

# Analyze a few key categorical features
key_categorical = ['Neighborhood', 'HouseStyle', 'SaleCondition']

for feature in key_categorical:
    if feature in df.columns:
        print(f"\nðŸ·ï¸  {feature.upper()}:")
        print("-" * 70)
        value_counts = df[feature].value_counts().head(10)
        for value, count in value_counts.items():
            percentage = (count / len(df) * 100)
            print(f"   {value:20s}: {count:4d} ({percentage:5.2f}%)")
```

```python
#  12: Analyze price by neighborhood
if 'Neighborhood' in df.columns:
    neighborhood_prices = df.groupby('Neighborhood')['SalePrice'].agg(['mean', 'median', 'count']).sort_values('mean', ascending=False)

    plt.figure(figsize=(16, 8))
    plt.barh(range(len(neighborhood_prices)), neighborhood_prices['mean'].values, color='coral')
    plt.yticks(range(len(neighborhood_prices)), neighborhood_prices.index)
    plt.xlabel('Average Sale Price ($)', fontsize=12)
    plt.title('Average House Price by Neighborhood', fontsize=14, fontweight='bold')
    plt.gca().invert_yaxis()

    # Add value labels
    for i, v in enumerate(neighborhood_prices['mean'].values):
        plt.text(v + 2000, i, f'${v:,.0f}', va='center', fontsize=9)

    plt.tight_layout()
    plt.show()

    print("\nðŸ˜ï¸  TOP 5 MOST EXPENSIVE NEIGHBORHOODS:")
    print("=" * 70)
    print(neighborhood_prices.head())
```

```python
#  13: Price trends by year built
if 'YearBuilt' in df.columns:
    yearly_prices = df.groupby('YearBuilt')['SalePrice'].agg(['mean', 'count'])

    plt.figure(figsize=(16, 6))
    plt.plot(yearly_prices.index, yearly_prices['mean'], linewidth=2, color='darkgreen', marker='o', markersize=4)
    plt.xlabel('Year Built', fontsize=12)
    plt.ylabel('Average Sale Price ($)', fontsize=12)
    plt.title('Average House Price by Year Built', fontsize=14, fontweight='bold')
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

    print("\nðŸ“… YEAR BUILT INSIGHTS:")
    print("=" * 70)
    print(f"â€¢ Oldest house: {df['YearBuilt'].min()}")
    print(f"â€¢ Newest house: {df['YearBuilt'].max()}")
    print(f"â€¢ Average year: {df['YearBuilt'].mean():.0f}")
```

```python
#  14: Generate comprehensive insights
print("=" * 80)
print(" " * 20 + "ðŸŽ¯ KEY INSIGHTS FROM ANALYSIS")
print("=" * 80)

print("\nðŸ“Š DATASET OVERVIEW:")
print(f"   â€¢ Analyzed {df.shape[0]:,} houses with {df.shape[1]} features")
print(f"   â€¢ Price range: ${df['SalePrice'].min():,.0f} - ${df['SalePrice'].max():,.0f}")
print(f"   â€¢ Average price: ${df['SalePrice'].mean():,.0f}")

print("\nðŸ”— STRONGEST PRICE PREDICTORS:")
top_5_corr = correlations.head(5)
for i, (feature, corr) in enumerate(top_5_corr.items(), 1):
    print(f"   {i}. {feature}: {corr:.3f} correlation")

print("\nâš ï¸  DATA QUALITY:")
if len(missing) > 0:
    print(f"   â€¢ {len(missing)} features have missing values")
    print(f"   â€¢ Top missing: {missing.index[0]} ({missing.iloc[0]} values)")
else:
    print("   â€¢ No missing values detected")

print("\nðŸ’¡ BUSINESS RECOMMENDATIONS:")
print("   1. Focus on overall quality and living area for price optimization")
print("   2. Location (neighborhood) significantly impacts pricing")
print("   3. Newer houses generally command higher prices")
print("   4. Garage and basement features are valuable")
print("   5. Consider data cleaning for features with high missing values")

print("\n" + "=" * 80)
print("âœ… EXPLORATORY DATA ANALYSIS COMPLETE!")
print("=" * 80)
```
